```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df= pd.read_csv("C:/Users/VINAY KUMAR PATEL/Downloads/machine learning/diabetes__1_.csv",header=0)
```

```python
df.head()
```

```python
df.describe()
```

```python
x=df.loc[:,df.columns!="Outcome"]
```

```python
x.head()
```

```python
y= df['Outcome']
```

```python
y.head()
```

# test train split

```python
from  sklearn.model_selection  import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

# tree making

```python
from sklearn import tree
cls_tree=tree.DecisionTreeClassifier(max_depth=3)
cls_tree.fit(x_train,y_train)
```

# predicting value

```python
y_train_pred= cls_tree.predict(x_train)
y_train_pred
```

```python
y_test_pred= cls_tree.predict(x_test)
y_test_pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
confusion_matrix(y_train , y_train_pred )
```

```python
confusion_matrix(y_test , y_test_pred )
```

```python
accuracy_score(y_train , y_train_pred )
```

```python
accuracy_score(y_test , y_test_pred )
```

# ploting decision tree

```python
dot_data = tree.export_graphviz(cls_tree,out_file = None,feature_names =x_train.columns,filled= True)
```

```python
from IPython.display import Image
```

```python
import pydotplus
```

```python
graph= pydotplus.graph_from_dot_data(dot_data)
```
